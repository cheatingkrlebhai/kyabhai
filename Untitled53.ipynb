```python
import pandas as pd
import numpy as np
import yfinance as yf
```

```python
stock = "AAPL"
start_date = "2020-01-01"
end_date = "2024-10-10"
```

```python
data = yf.download(stock, start=start_date, end=end_date)
data
```

```python
short_window = 50
long_window = 200
data["SMA50"] = data["Close"].rolling(window=short_window).mean()
data["SMA200"] = data["Close"].rolling(window=long_window).mean()
```

```python
data["Signal"] = 0
data["Signal"][short_window:] = np.where(data["SMA50"][short_window:]>data["SMA200"][short_window:],1,0)
data["Position"] = data["Signal"].diff()
data
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(14,7))
plt.plot(data["Close"], label = "Close Price", alpha = 0.5)
plt.plot(data["SMA50"], label = "SMA50", alpha=0.75)
plt.plot(data[data["Position"]==1].index, data["SMA50"][data["Position"]==1], "^", markersize=10, color="g", label="Buy Signal")
plt.plot(data[data["Position"]==-1].index, data["SMA50"][data["Position"]==-1], "v", markersize=10, color="r", label="Sell Signal")
plt.title("MV")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend(loc="upper left")
plt.show()
```

```python
window = 20
threshold = 2
```

```python
symbol = "AAPL"
start_date = "2020-01-01"
end_date = "2024-10-10"
data1 = yf.download(symbol, start=start_date, end=end_date)
data1
```

```python
data1["Close"].rolling(window=window).mean()
```

```python
data1["Signal"] = np.where(data1["Close"] < data1["Lower"], 1, 0)
```

```python
data_2 = data1.stack().reset_index(drop = False)
```

```python
data_2 = data_2.set_index('Date')
```

```python
data_2.columns.name = None
```

```python
data_2["MA"] = data_2["Close"].rolling(window=window).mean()
data_2["STD"] = data_2["Close"].rolling(window=window).std()
data_2["Upper"] = data_2["MA"] + (threshold * data_2["STD"])
data_2["Lower"] = data_2["MA"] - (threshold * data_2["STD"])
data_2
```

```python
data1
```

```python
data_2["Signal"] = np.where(data_2["Close"] < data_2["Lower"], 1, 0)
data_2["Signal"] = np.where(data_2["Close"] > data_2["Upper"], -1, data_2["Signal"])
data_2
```

```python
data_2["Position"]=data_2["Signal"].diff()
data_2
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(14,7))
plt.plot(data_2["Close"], label = "Close Price", alpha = 0.5)
plt.plot(data_2["MA"], label = "MA", alpha=0.75)
plt.figure(figsize=(14,7))
plt.plot(data_2["Close"], label = "Close Price", alpha = 0.5)
plt.plot(data_2["MA"], label = "MA", alpha=0.75)
plt.plot(data_2["Upper"], label = "Upper", linestyle = "--", alpha=0.75)
plt.plot(data_2["Lower"], label = "Lower", linestyle = "--", alpha=0.75)
plt.plot(data_2[data_2["Position"]==1].index, data_2["Close"][data_2["Position"]==1],"^", markersize=12, color="g", label="Buy Signal")
plt.plot(data_2[data_2["Position"]==-1].index, data_2["Close"][data_2["Position"]==-1],"v", markersize=12, color="r", label="Sell Signal")
plt.title("MV")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend(loc="upper left")
plt.show()
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend(loc="upper left")
plt.show()
```

```python
symbol = "AAPL"
start_date = "2020-01-01"
end_date = "2024-10-10"
data3 = yf.download(symbol, start=start_date, end=end_date)
data3
rsi_buy_threshold = 30
rsi_sell_threshold = 70
```

```python
delta = data3["Close"].diff()
gain = (delta.where(delta > 0, 0)).fillna(0)
loss = (-delta.where(delta < 0, 0)).fillna(0)
rs = gain/loss
data3["rsi"] = 100 - (100/(1+rs))
```

```python
data3["Signal"] = np.where(data3["rsi"]<rsi_buy_threshold,1,0)
data3["Signal"] = np.where(data3["rsi"]>rsi_sell_threshold,-1,data3["Signal"])
data3["Position"] = data3["Signal"].diff()
data3
```

```python
plt.figure(figsize=(14,7))
plt.plot(data3["Close"],label="Close Price", alpha=0.5)
plt.plot(data3[data3["Position"]==1].index,data3["Close"][data3["Position"]==1],"^",markersize=12,color="g",label="Buy Signal")
plt.plot(data3[data3["Position"]==-1].index,data3["Close"][data3["Position"]==-1],"v",markersize=12,color="r",label="Sell Signal")
plt.title(f"RSI Trading Strategy for {symbol}")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()
plt.show()
```
